```python
import pandas as pd
import numpy as np
```

```python
url_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx"
```

```python
df = pd.read_excel(
    url_path, 
    sheet_name="Canada by Citizenship",
    skiprows=range(20),
    skipfooter= 2
)                 
```

```python
df.columns
```

```python
df.dtypes
```

```python
country_group = df.groupby(['OdName'], as_index = True).sum()
```

```python
country_group.head()
```

```python
country_group.drop('DEV',axis=1, inplace=True)
country_group.drop('AREA',axis=1, inplace=True)
country_group.drop('REG',axis=1, inplace=True)
```

```python
total = country_group.iloc[:,5:].sum(axis=1)
```

```python
country_group['total'] = total
```

```python
country_group.head()
```

```python
country_group.to_csv('total_population_by_country.csv', index=False)
```
