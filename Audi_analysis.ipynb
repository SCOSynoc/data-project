```python
import pandas as pd
import numpy as np
```

```python
path='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv'
```

```python
df = pd.read_csv(path)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
sns.regplot(x="length", y="price", data=df)
plt.ylim(0,)
```

# As length of car increses the car price increases

```python
sns.regplot(x="width", y="price", data=df)
plt.ylim(0,)
```

# As width increases the car price also increases

```python
sns.regplot(x="curb-weight", y="price", data=df)
plt.ylim(0,)
```

# As curb weight increases the price increases

```python
sns.regplot(x="horsepower", y="price", data=df)
plt.ylim(0,)
```

# As horse power increases the price increases

```python
sns.regplot(x="city-mpg", y="price", data=df)
plt.ylim(0,)
```

# Here as city mpg increases the price decreases

```python
sns.regplot(x="highway-mpg", y="price", data=df)
plt.ylim(0,)
```

# Here as highway-mpg increases the price decreases

```python
sns.regplot(x="bore", y="price", data=df)
plt.ylim(0,)
```

# As bore increases the price value increases

```python
sns.regplot(x="wheel-base", y="price", data=df)
plt.ylim(0,)
```

# As wheel-base increases the car price increases

# So based on the above analysis if customer wants to buy an audi with 
##### Higher length 
##### Higher curb weight 
##### Higher horsepower 
##### lower highway-mpg 
##### lower city-mpg
